```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import seaborn
import os
import random
```

## Data Analizi

```python
train_file='/content/drive/MyDrive/Python/calismalarim/deep_learning/man_woman/dataset/train/'
test_file='/content/drive/MyDrive/Python/calismalarim/deep_learning/man_woman/dataset/test/'
categories1=['man','woman']
deneme=[]

for category in categories1:
  class_num=categories1.index(category)
  for i in os.listdir(train_file+category):
    img_array=cv2.imread(train_file+category+'/'+i,cv2.IMREAD_GRAYSCALE)
    new_array=cv2.resize(img_array,(50,50))
    deneme.append([new_array,class_num])
    print(img_array.shape)
    print(new_array.shape)
    plt.imshow(new_array)
    break

```

```python
def load_data(file_path,data,categories):

  for category in categories:
    class_num=categories1.index(category)
    for i in os.listdir(file_path+category):
      img_array=cv2.imread(file_path+category+'/'+i,cv2.IMREAD_GRAYSCALE)
      new_array=cv2.resize(img_array,(50,50))
      data.append([new_array,class_num])
```

```python
train_data=[]
test_data=[]
```

```python
load_data(train_file,train_data,categories1) 
load_data(test_file,test_data,categories1) 
```

```python
import random

random.shuffle(train_data)
random.shuffle(test_data)
```

```python
for sample in train_data[:10]:

  print(sample[1])
```

## Train-Test (x_train, y_train)

```python
x_train=[]
y_train=[]
x_test=[]
y_test=[]
```

```python
for features,label in train_data:

  x_train.append(features)
  y_train.append(label)

for features1,label1 in test_data:

  x_test.append(features1)
  y_test.append(label1)
```

```python
print(len(x_train))
print(len(y_train))
```

## Datayı kaydetme (Save)

```python
import pickle


pickle_out = open("X.train","wb")
pickle.dump(x_train, pickle_out)
pickle_out.close()

pickle_out = open("Y.train","wb")
pickle.dump(y_train, pickle_out)
pickle_out.close()


pickle_out = open("X.test","wb")
pickle.dump(x_test, pickle_out)
pickle_out.close()

pickle_out = open("y.test","wb")
pickle.dump(y_test, pickle_out)
pickle_out.close()
```
